## Exercise Set 2: Reordering Columns

```python
import pandas as pd

data = {
"Name": ["Alice", "Bob", "Charlie"],
"Dept": ["HR", "Finance", "IT"],
"Sal": [60000, 70000, 80000],
}

df = pd.DataFrame(data)
```

## Exercise 1

```python
df.rename(columns={
    'Dept': 'Department',
    'Sal': 'Salary'
}, inplace=True)
print("1. Renamed 'Dept' and 'Sal':")
print(df)
```

## Exercise 2

```python
df = df[['Name', 'Department','Salary']]
cols = df.columns.tolist()
new_order = [cols[-1]] + cols[:-1]  
df = df[new_order]
```

## Exercise Set 3: GroupBy and Agg

```python
import pandas as pd

data = {
"Store": ["A", "A", "A", "B", "B", "C", "C", "C", "C"],
"Product": ["Apples", "Oranges", "Bananas", "Apples", "Bananas", "Oranges", "Apples", "Bananas", "Bananas"],
"Sales": [100, 80, 90, 120, 110, 60, 95, 85, 75],
"Profit": [30, 25, 20, 40, 35, 15, 32, 22, 18],
}

df = pd.DataFrame(data)
print(df.head())
```

## Exercise 1

```python
df.groupby('Store').agg({
    'Sales': 'mean',
    'Profit': 'mean'
})
```

## Exercise 2

```python
df.groupby('Product').agg({
    'Sales': 'max',
    'Profit': 'min'
})
```

## Exercise 3

```python
df.groupby('Store').agg(
    Avg_Sales=('Sales', 'mean'),
    Max_Profit=('Profit', 'max')
)
```

## Exercise 4

```python
df.groupby(['Store', 'Product']).agg({'Sales': 'sum'})
```
